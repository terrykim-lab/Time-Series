**1. 프로젝트 개요**



본 프로젝트는 자율주행 차량의 사고 데이터를 다변량 시계열 분석 기법을 통해 분석함으로써, 사고 발생의 시간적 패턴과 관련 요인을 규명하는 것을 목표로, 이를 통해 자율주행 기술의 안전성 향상과 교통 정책 수립에 기초 자료를 제공하고자 함.



---



**2. 활용 데이터셋**


Kaggle을 이용하여 아래의 공개 데이터셋들을 확보하였으며, 자율주행 및 일반 차량의 사고 데이터를 중심으로 분석할 예정.


1. [Automated Driving Crashes](https://www.kaggle.com/datasets/michaelbryantds/automated-driving-crashes)

   ⇒ 자율주행차의 사고 사례, 발생 일시, 위치, 유형 등


2. [Traffic Accidents Dataset](https://www.kaggle.com/datasets/oktayrdeki/traffic-accidents)

   ⇒ 다양한 지역의 교통사고 통계 정보, 날씨, 도로 상황 등

3.	[Udacity Self-Driving Car Dataset](https://www.kaggle.com/datasets/sshikamaru/udacity-self-driving-car-dataset)

   ⇒ 자율주행 차량 센서 기반 주행 정보 및 사고 관련 데이터 등



---


**3. 분석 목적**

⇒ 자율주행 차량의 사고 발생 비율이 시간에 따라 어떻게 변화하는지 파악

⇒ 사고에 영향을 미치는 외부 요인(날씨, 시간대, 도로 상태 등)을 분석

⇒ 사고 발생 예측 가능성 탐색을 통해 기술 및 정책 개선 방향 제안






---



**4. 분석 흐름 (계획)**

1. 데이터 수집 및 전처리

  ⇒ 확보한 3종 데이터셋 정리 및 병합
  
  ⇒ 시계열 분석을 위한 날짜/시간 기준 정규화, 결측치 확인

2. 탐색적 데이터 분석 (EDA)

  ⇒ 변수 간 분포 확인 및 시각화

  ⇒ 상관관계 탐색, 시계열 특성 점검

3. 시계열 모델링 및 예측

  ⇒ Vector Autoregression(VAR), Dynamic Factor Model 적용 예정
  
  ⇒ 필요 시 단변량 ARIMA 모델 비교

4. 사고 패턴 및 시사점 도출

  ⇒ 사고 증가 조건 분석 및 예측력 검증

  ⇒ 자율주행 차량의 위험 요인 도출
  





---



**5. 진행 현황**


⇒ 주요 데이터셋 3종 확보 완료

⇒ 변수 구성 및 포함 항목에 대한 1차 검토 진행

⇒ 시게열 분석을 위한 구조화(날짜 변수 정리, 사고 발생 단위 확인 등) 준비 중


¤ 전처리 및 분석 작업은 중간보고서 제출 이후 진행 예정





---



**6. 기대 효과**


⇒ 자율주행 차량 사고의 시계열적 변화양상에 대한 데이터 기반 인사이트 확보

⇒ 사고에 영향을 주는 요인을 정량적으로 도출

⇒ 사고 예방 및 기술개선, 정책 수립을 위한 근거 자료 제공
